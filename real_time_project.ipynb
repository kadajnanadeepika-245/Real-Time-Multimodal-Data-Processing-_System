```python
from transformers import pipeline
```

```python
summarize = pipeline("summarization")
```

```python
text = input('')
```

```python
result = summarize(text,max_length= 100)
```

```python
result
```

```python
q_a = pipeline('question-answering')
```

```python
while True:
    question = input('Enter question (type "exit" to quit): ')
    if question.lower() == 'exit':
        break

    result = q_a(
        question=question,
        context=text
    )

    print(result)
```
